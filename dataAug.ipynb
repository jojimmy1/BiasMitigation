```python
import numpy
import pandas
from sklearn.preprocessing import LabelEncoder
```

```python
kiva_label = pandas.read_csv('/home/li3975/cs587/data/kiva/kiva_loans.csv')

kiva_label = kiva_label.drop(columns=['id',  'activity', 'country', 'posted_time', 
                          'disbursed_time', 'funded_time', 'tags', 'use', 
                          'region', 'date'])
kiva_label = kiva_label.dropna()
```

```python
kiva_label = kiva_label[~kiva_label['borrower_genders'].str.contains(',', na=False)]

print(kiva_label.shape)

le = LabelEncoder()
kiva_label['borrower_genders'] = le.fit_transform(kiva_label['borrower_genders'])
# mapping = dict(zip(le.classes_, range(1, len(le.classes_)+1)))
# print(len(mapping))
# print(mapping['male'])
kiva_label['country_code'] = le.fit_transform(kiva_label['country_code'])
kiva_label['currency'] = le.fit_transform(kiva_label['currency'])
kiva_label['repayment_interval'] = le.fit_transform(kiva_label['repayment_interval'])
kiva_label['sector'] = le.fit_transform(kiva_label['sector'])
```

```python
print(kiva_label.shape)
kiva_label
```

```python
kiva_label.to_csv('/home/li3975/cs587/dataAug/kiva_label.csv', index=False)
```

```python
kiva_onehot = pandas.read_csv('/home/li3975/cs587/data/kiva/kiva_loans.csv')

kiva_onehot = kiva_onehot.drop(columns=['id',  'activity', 'country', 'posted_time', 
                          'disbursed_time', 'funded_time', 'tags', 'use', 
                          'region', 'date'])
kiva_onehot = kiva_onehot.dropna()
kiva_onehot.shape
```

```python
studentperf_label = pandas.read_csv("/home/li3975/cs587/data/student-performance-factors/StudentPerformanceFactors.csv")
studentperf_onehot = pandas.read_csv("/home/li3975/cs587/data/student-performance-factors/StudentPerformanceFactors.csv")
```

```python
studentperf_label.drop(columns={'Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities', 'Internet_Access', 'School_Type', 'Peer_Influence'}, inplace = True)
studentperf_onehot.drop(columns={'Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities', 'Internet_Access', 'School_Type', 'Peer_Influence'}, inplace = True)
```

```python
studentperf_label.dtypes
```

```python
import pandas as pd
#create naive encoding:
studentperf_label = studentperf_label.dropna()
studentperf_onehot = studentperf_onehot.dropna()
le = LabelEncoder()

cat_columns = []
for column in studentperf_label.select_dtypes(include=['object']).columns:
    studentperf_label[column] = le.fit_transform(studentperf_label[column])
    cat_columns.append(column)

#create one hot
studentperf_onehot = pd.get_dummies(studentperf_onehot, columns=cat_columns)
```
