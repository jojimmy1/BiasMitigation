```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
techniques = ['Pre-Processing', 'In-processing', 'Post-processing']
metrics = ['Accuracy', 'Dao', 'Deo', 'F_Dao', 'F_Deo', 'CV']
```

```python
label_encoding_kiva = [
    [0.91, 0.04591836735, 0.07142857143, 0.8682142857, 0.845, 0.2230378317],  # Pre-Processing
    [0.9477, 0.0089, 0.012, 0.9363, 0.9393, 0.1743],                         # In-processing
    [0.8646157889, 0.00439471, 0.006515324, 0.8585336392, 0.8567049844, 0.168472409]  # Post-processing
]

label_encoding_sp_family_income  = [
    [0.9, 0.05847953216, 0.005847953216, 0.8473684211, 0.8947368421, 0.0007087172218],
    [0.5549, 0, 0, 0.5549, 0.5549, 0],
    [0.8119447087, 0.14196907, 0.085705817, 0.6966736735, 0.7423563241, 0.030741692]
]

```

```python
def compare_metrics(df):
    x = np.arange(len(metrics))  # Label locations
    width = 0.25  # Width of each bar

    fig, ax = plt.subplots(figsize=(12, 6)) 

    # Create bars for each technique
    for i, technique in enumerate(techniques):
        ax.bar(x + i * width, df[i], width, label=technique)

    # Labels, title, and legend
    ax.set_xlabel('Metrics', fontsize=12)
    ax.set_ylabel('Values', fontsize=12)
    ax.set_title('Comparison of Techniques Across Metrics', fontsize=14)
    ax.set_xticks(x + width)
    ax.set_xticklabels(metrics, rotation=45, fontsize=10)
    ax.legend(title="Techniques", fontsize=10)

    # Add value annotations
    for i, technique_data in enumerate(df):
        for j, value in enumerate(technique_data):
            ax.text(x[j] + i * width, value + 0.005, f'{value:.3f}', ha='center', va='bottom', fontsize=8)

    # Display plot
    plt.tight_layout()
    plt.show()
```

```python
compare_metrics(label_encoding_kiva)
```

```python
compare_metrics(label_encoding_sp_family_income)
```
